```python
import pandas as pd
```

```python
tabela = pd.read_csv("brasileirao_serie_a.csv")
display (tabela)
```

```python
# Qual o número medio de faltas por jogo
faltas_totais= tabela['faltas_mandante'] + tabela['faltas_visitante']
faltas_media = faltas_totais.sum() / 8079

print (faltas_media)
```

```python
# Qual é o time com mais gol 
mandante_melhor = tabela.groupby('time_mandante')['gols_mandante'].sum()
visitante_melhor = tabela.groupby('time_visitante')['gols_visitante'].sum()
o_melhor = mandante_melhor.add(visitante_melhor, fill_value=0)
o_melhor = o_melhor.idxmax()


print(o_melhor)
```

```python
# Quais jogos atingiram seu publico maximo ?

publico_maximo = tabela['publico'].idxmax()
jogos_maximos = tabela.loc()[publico_maximo]

print(jogos_maximos)
```

```python
# Quais são os melhores jogos do São Paulo ? 

# Filtrando os jogos em que o São Paulo participou
jogos_sp_mandante = tabela[tabela['time_mandante'] == "São Paulo"]
jogos_sp_visitante = tabela[tabela['time_visitante'] == "São Paulo"]
jogos_sp = pd.concat([jogos_sp_mandante, jogos_sp_visitante])

# Calculando o total de gols do São Paulo em cada jogo
jogos_sp['total_gols_sp'] = jogos_sp['gols_mandante'].where(jogos_sp['time_mandante'] == 'São Paulo', jogos_sp['gols_visitante'])


jogos_sp = jogos_sp.sort_values(by='total_gols_sp', ascending=False)

display(jogos_sp.head(10))
```

```python
# Quem ganhou o Campeonato em 2008 ?

jogos_2008 = tabela[tabela['ano_campeonato'] == 2008].copy()

jogos_2008['pontuacao'] = 0

# Calculando os pontos 
jogos_2008.loc[jogos_2008['gols_mandante'] > jogos_2008['gols_visitante'], 'pontuacao'] +=3
jogos_2008.loc[jogos_2008['gols_mandante'] < jogos_2008['gols_visitante'], 'pontuacao'] +=0
jogos_2008.loc[jogos_2008['gols_mandante'] == jogos_2008['gols_visitante'], 'pontuacao'] +=1

pontuacao_total_2008 = jogos_2008.groupby('time_mandante')['pontuacao'].sum().reset_index()

# Ordenar os times com base na pontuação
classificacao_2008 = pontuacao_total_2008.sort_values(by='pontuacao', ascending=False).reset_index(drop=True)

print(classificacao_2008)
```

```python
# Qual time visitante mais ganhou jogos ? 

vitorias_visitante = tabela[tabela['gols_visitante'] > tabela['gols_mandante']]
vitorias_por_time = vitorias_visitante['time_visitante'].value_counts()

time_mais_vitorioso = vitorias_por_time.idxmax()
numero_vitorias = vitorias_por_time.max()

print(f"O time visitante que mais ganhou jogos foi '{time_mais_vitorioso}' com um total de {numero_vitorias} vitórias.")
```

```python
# Detalhes estatisticos dos gols dos times mandantes 

# Calcular detalhes estatísticos dos gols dos times mandantes
estatisticas_gols_mandantes = tabela['gols_mandante'].describe()

print("Detalhes Estatísticos dos Gols dos Times Mandantes:")
print(estatisticas_gols_mandantes)
```
